```python
from pykrx import stock
import datetime

# 시작일과 종료일 설정 (최근 1년)
start_date = datetime.datetime.now() - datetime.timedelta(days=365)
end_date = datetime.datetime.now()

# 코스피 지수 코드 설정
code = "1001"

# 코스피 지수 데이터 가져오기
df = stock.get_index_ohlcv_by_date(start_date.strftime('%Y%m%d'), end_date.strftime('%Y%m%d'), code)

df
```

```python
df['수익률'] = df["종가"].pct_change()
df[['수익률']]
```

```python
df[['수익률']].mean() * np.sqrt(252)
```

```python
df[['수익률']].std() * np.sqrt(252)
```

```python
df["종가"].plot()
```

```python
from scipy import stats

statistic, pvalue = stats.shapiro(df['종가'])
print("Shapiro-Wilk test statistic : {}".format(statistic))
print("p-value : {}".format(pvalue))
```
