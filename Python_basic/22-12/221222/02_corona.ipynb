```python
# 복습
# 코로나 파일 로드
# 결측치 존재 여부 확인 
# unnamed:0 컬럼 삭제
# 컬럼의 이름 변경 [등록일시. 사망자, 확진자, 게시글번호, 기준일, 기준시간, 수정일시, 누적 의심자, 누적 확진률]
# 등록일시 기준 오름차순 정렬 인덱스 리셋 기존의 인덱스 삭제
# 새로운 파생변수 일일 사망자 일일 화진자 생성 
# 게시글번호, 기준일, 기준시간, 컬럼 삭제
```

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("../csv/corona.csv")
df = pd.DataFrame(df)
df.head(3)
```

```python
df.isna().sum()
```

```python
df.drop(["Unnamed: 0"], axis=1, inplace = True)
```

```python
df.columns = ["등록일시", "사망자", "확진자","게시글번호", "기준일", "기준시간", "수정일시", "누적의심자", "누적확진율"]
```

```python
df.sort_values(["등록일시"], inplace=True)
```

```python
df.reset_index(drop =True, inplace=True)
```

```python
df["일일확진자"] = df["확진자"] - df.shift(1)["확진자"]
```

```python
df["일일사망자"] = df["사망자"] - df.shift(1)["사망자"]
```

```python
df.drop(["게시글번호" ,"기준일", "기준시간"],axis=1,inplace=True)
```

```python
error = df[(df["일일사망자"] < 0)].index
```
