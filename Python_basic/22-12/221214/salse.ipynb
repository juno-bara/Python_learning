```python
import pandas as pd
```

```python
# csv 폴더에 있는 salse record.csv 파일로드
# Order ID, Total, Revenue, Total Cost, Total profit 컬럼을 삭제
# 컬럼의 이름 변경[권역, 국가, 상품종류, 채널, 우선순위, 주문일자, 발송일자, 
# 단가, 원가]
# 권역 국가를 기준으로 오름차순 정열
# index 값초기화하고 기존의 인덱스 삭제
 
```

```python
sales = pd.read_csv("../csv/Sales Records.csv")
sales_df = sales.copy()
sales_df
```

```python
# Order ID, Total, Revenue, Total Cost, Total profit 컬럼을 삭제
sales_df = sales_df.drop(["Order ID","Total Revenue","Total Cost", "Total Profit"],axis=1)
```

```python
sales_df
```

```python
#컬럼의 이름 변경[권역, 국가, 상품종류, 채널, 우선순위, 주문일자, 발송일자, 판매단위
# 단가, 원가]
sales_df.columns = ["권역","국가","상품종류","채널", "우선순위", "주문일자", "발송일자","판매단위","단가","원가"]
```

```python
sales_df.head()
```

```python
# 권역 국가를 기준으로 오름차순 정열
sales_df = sales_df.sort_values(["권역","국가"])
```

```python
sales_df
```

```python
# index 값초기화하고 기존의 인덱스 삭제
# reset_index(drop=false) >> 기본값 drop=true 로 바꾸면 기존의 인덱스 삭제
sales_df = sales_df.reset_index(drop=True)
```

```python
sales_df.head()
```

```python
# reset_index() 안에 drop은 기본값이 false
# drop 매개변수의 인자 값이 True가 되면 기존의 인덱스를 삭제하고 
# index를 초기화 한다

sales_df.sort_values(["국가"]).reset_index(drop=True)
```

```python
#단가의 컬럼의 float형태의 데이터를 반올림 > int형태로 변경
# 반올림 함수 > round(n)
# n의 값은 소수점의 자릿수
# 데이터의 타입 변경하는 함수 >> astype("int")


sales_df["단가"] = sales_df["단가"].round(0).astype("int")
```

```python
sales_df.head(6)
```

```python
# 원가 컬럼의 데이터를 반올림하지 않고 int 타입으로 변경

sales_df["원기"] = sales_df["원가"].astype("int")
```

```python
sales_df.head(6)
```

```python
# 새로운 파생변수 (총 단가, 총 원가, 총 이윤)생성
# 1. 총단가 = 단가 * 판매단위
# 2. 총원가 = 원가 * 판매단위
# 3. 총이윤 = 총단가 - 총원가
```

```python
# sales["총 단가"]

sales_df["단가"] * sales_df["판매단위"]
```

```python
sales_df["총 단가"] = sales_df["단가"] * sales_df["판매단위"]
sales_df["총 원가"] = sales_df["원가"] * sales_df["판매단위"]

sales_df["총 이윤"] = (sales_df["총 단가"] - sales_df["총 원가"])
sales_df
```

```python
sales_df.head(6)
```

```python
# loc[]
# 특정 행이나 특정 열의 위치를 추출할때 사용
# loc[행의위치, 열의 위치]

# sales 데이터에서 인덱스가 11부터 마지막까지 ( = 0부터 10까지는 삭제) 
sales_df.loc[11:]
```

```python
sales_df.loc[11:,"권역":"발송일자"]
```

```python
# 해당하는 컬럼의 데이터를 그룹화하여 카운트 체크
sales_df["채널"].value_counts() 
```

```python
# 국가별로 그룹화하여 총 이윤의 합계을 구한 다음 가장 이윤이 
    # 높은 나라의 top 10을 뽑으시오

# 1 국가 별로 그룹화
# 2 총 이윤의 합계 > 데이터프레임
# 3 총 이윤을 가지고 내림차순으로 정렬
# 4 상위 10개를 출력

# 그룹화 함수 >> groupby("기준이 되는 컬럼")

test = sales_df.groupby("국가").sum()
# 국가를 기준으로 그룹화 하여 컬럼의 합을 구하는 식
```

```python
test[["총 이윤"]]
#test["총 이윤"]
```

```python
# 총이윤을 기준으로 내림차순 정렬
# sort.values() > 기본 값은 오름차순으로 정렬
# 내림차순 정렬로 변경하려면 함수의 매개변수 asceding

test.sort_values(["총 이윤"], ascending = False).head(10)
```
