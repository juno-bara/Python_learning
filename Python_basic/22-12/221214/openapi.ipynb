```python
import requests
from urllib import parse
import xmltodict
import pandas as pd
```

```python
# open api 이용하여 서버와의 통신을 하여 데이터를 가져오는 작업

# 어느 서버에서 데이터를 가져와야 하나 

url = "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson"

# 공공데이터 포털에서 주소와 함께 request parameter 
# 요청변수 > 
# 서비스 키
serviceKey = "dtbWOdJ%2FCz5HE0DGLU%2BCRPe7pOW0NIQBUcGEqsHZaTRiYCI%2F5%2BzugwzQjcvuId7NPdg6rUiW%2Bft3fm7yqyD4pw%3D%3D"

# 페이지 번호 
# 페이지 당 데이터의 수
# 시작 시간
# 종료 시간 

param = { "serviceKey" : parse.unquote(serviceKey),
    "pageNo" : "1",
    "numOfRows" : "5000",
    "startCreateDt" : "20200101",
    "endCreatDt" : "20221214" 
}

# 이제 여기부터 통신 시작
reponse = requests.get(url,params = param)
# reponse의 데이터의 형태는 xml

print(reponse)
# 200 > 정상,  404 > 통신에러 

# xml데이트를 딕셔너리 형태로 변환
result = xmltodict.parse(reponse.content)

# print(result)












```

```python

rows = result["response"]["body"]["items"]["item"]
# print(rows)
```

```python
corona_df = pd.DataFrame(rows)
corona_df.head(10)
```

```python
# openapi를 이용하여 받아온 데이터를  csv파일로 저장

corona_df.to_csv("corona.csv",encoding="UTF-8")
```
