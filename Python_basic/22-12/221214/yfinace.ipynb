```python
AAPL = yf.download("AAPL", start="2020-01-01")
```

```python
AAPL.head(6)
```

```python
# 소숫점 두 번재 자리까지만 두고 반올림

AAPL[["Open", "High", "Low", "Close"]] = \
    AAPL[["Open", "High", "Low", "Close"]].round(2)
```

```python
AAPL.isna().sum()
```

```python
# 등락폭 , 등락율 파생변수 생성
# 등락폭 = 증가 - 시가
# 등락율 = 등락폭 / 시가 * 100

AAPL["등락폭"] = AAPL["Open"] - AAPL["Close"]
AAPL["등락율"] = AAPL["등락폭"] / AAPL["Open"] * 100
AAPL.head(6)
```

```python
# 시각화 

import matplotlib.pyplot as plt
```

```python
plt.plot(AAPL["등락율"].values)
plt.show()
```

```python
# 그래프 4개 출력 
# subplot(행의수, 열의수, 선택할 영역)
# 영역을 나누는 작업 > 나눈 영역에서 영역 선택
# ex) subplot(2,2,1) > 4개 구역으로 나누 후 첫 번째 구역 선택

# 영역 분할 및 선택 
plt.subplot(4,1,1)
#그래프 그리기
plt.plot(AAPL["Open"].values)

plt.subplot(4,1,2)
plt.plot(AAPL["Close"].values)

plt.subplot(4,1,3)
plt.plot(AAPL["등락율"].values)

plt.subplot(4,1,4)
plt.plot(AAPL["Volume"].values)
```
