```python
import pandas as pd
```

```python
# 코로나 데이터 로드

corona_df = pd.read_csv("corona.csv")
corona_df = pd.DataFrame(corona_df)
corona_df
```

```python
# 데이터 프레임에서 컬럼 삭제
# drop() 함수 이용하여 컬럼을 삭제
# drop() 함수는 axis하는 매개변수를 기준으로 행, 열을 삭제한다
# inplace라는 매개변수 기본값이 false >>> true로 변경하면 
# 기준이 되는 데이터 값이 변환

corona_df.drop(["Unnamed: 0"],axis=1, inplace= True)
# corona = corona_df.drop(["Unnamed: 0"], axis=1))
# corona_df = corona_df.drop(["Unnamed: 0"],axis=1)
# 셋 결과가 같음


```

```python
# 컬럼의 이름을 변경
# 데이터 프레임에 있는 컬럼의 값들을 출력하는 방법
# 데이터프레임명.colunms >> 컬럼의 값 산출 
 
corona_df.columns 
```

```python
corona_df.columns  = ["등록일시", "사망자", "확진자", "게시글 번호", 
                "기준일", "기준시간", "수정일시", "누적 의심자", "누적 확진률"]
```

```python
corona_df.head(6)
```

```python
# corona_df 데이터 프레임에서 게시글 번호 삭제
# corona_df 라는 변수를 수정
 
# case_1 
# corona_df_1 = corona_df.drop(["게시글 번호"], axis=1)
# case_2 
corona_df.drop(["게시글 번호"], axis=1,inplace=True)
```

```python
corona_df.head(6)
```

```python
# 데이터 프레임의 정렬 변경

corona_df.sort_values(["등록일시"],inplace=True)
```

```python
corona_df.head(6)
```

```python
# 인덱스를 초기화하여 새로운 넘버링 

# corona_df.reset_index().drop("index",axis=1)
corona = corona_df.reset_index(drop=True)
# corona_df.reset_index(drop=True, inplace = true)
```

```python
corona.tail(6)
```

```python
# 데이터간의 차이를 나타내는 함수 
# diff()
# corona["확진자"].diff()

# corona 데이터프레임에 새로운 파생변수 (일일 확진자)추가
corona["일일 확진자"] = corona["확진자"].diff() 
```

```python
corona.head(6)
```

```python
# 새로운 파생변수 일일 사망자 추가 
# 누적 의심자, 누적 확진률 삭제 (corona 데이터 프레임 변경)

corona["일일 사망자"] = corona["사망자"].diff()

corona = corona.drop(["누적 의심자"],axis=1)
corona = corona.drop(["누적 확진률"],axis=1)

# corona.drop(["누적 의심자,"누적 확진률"],axis=1, inplace = true)

```
