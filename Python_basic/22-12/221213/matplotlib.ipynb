```python
import matplotlib.pyplot as plt
```

```python
# plot 그래프 >> 라인 그래프 생성

plt.plot([1,2,3,4])

# 도회지를 그려준다
plt.show()
```

```python
plt.plot([1,2,3,4],[1,200,3,500])
plt.show()
```

```python
x = [1,2,3,4]
y = [5,10,8,9]

# plot(x축의 값, y축의 값, 옵션)
# 옵션 선의 종류 선의 색

plt.plot(x,y)
plt.show()
```

```python
y2 = [4,2,5,1]

#첫번째 라인의 값
plt.plot(x,y,"--y")

#두번쩨 라인의 값
plt.plot(x,y2,".g")

# x축으 이름과
plt.xlabel("class")

# y 축의 이름 
plt.ylabel("count")

# 그래프의 title
plt.title("class & count")

# 범례의 추가
plt.legend(["A","B"])

plt.show()
```

```python
# subplot() > 여러 그래프를 그릴 때 영역을 나누고 영역을 지정
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,y2)

plt.show
```

```python
# 바형 그래프

x = [1,2,3,4]
y = [80,60,90,70]

plt.bar(x,y)
plt.show()
```

```python
# 막내너비 변경, 막대의 색 변경
plt.bar([1,2,3,4], y, color=["g","r","b","m"], width=0.8)


# x축 이름 변경
# plt.xticks(x,["maths","science","history","english"])
plt.xticks(x, ["A", "B", "C", "D"])
plt.show()
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.show()
```

```python
## 산점도 그래플에서 옵션값 
# 영역 확장, 색상 변경, 투명도 조절
color = np.random.rand(50)
area = (20*np.random.rand(50))**2

plt.scatter(x,y,s=area,c=color,alpha=0.7)
plt.show()
```
