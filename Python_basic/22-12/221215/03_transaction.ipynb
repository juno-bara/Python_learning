```python
# 데이터의 결합2

# 1. customer_master, item_master 파일 로드
# 2.두 개의 데이터를 tr_mg에 조건에 맞게 조인 결합 
# 3. 결합이 된 데이터에서 quantity 컬럼과 item_price컬럼을 가지고 total_price 라는
#   파생변수 생성
# payment_date 를 시계열 데이터로 변경
```

```python
cm = pd.read_csv("../csv/customer_master.csv")
im = pd.read_csv("../csv/item_master.csv")
```

```python
cm
```

```python
m = pd.merge(tr_mg, cm, on = "customer_id", how = "left")
m.head(2)
```

```python
im
```

```python

tr_mg_2 = pd.merge(m,im,on="item_id",how="left")
```

```python
tr_mg_2.isna().sum()
```

```python
tr_mg_2.head(3)
```

```python
# 결합이 된 데이터에서 quantity 컬럼과 item_price컬럼을 가지고 total_price 라는 파생변수 생성


tr_mg_2["total_price"] = (tr_mg_2["item_price"] * tr_mg_2["quantity"])
```

```python
#  payment_date를 시계열 데이터로 변경

tr_mg_2["payment_date"] = pd.to_datetime(tr_mg_2["payment_date"],format="%Y-%m-%d")
```

```python
tr_mg_2.loc[tr_mg_2["quantity"] > 2]
```

```python
tr_mg_2.loc[tr_mg_2["price"] != tr_mg_2["total_price"]]
```

```python
# 시계열데이터 이용한는 예제
# 시계열데이터에서 년, 월 추출해서 새로운 파생변수 생성
# paymaet_month 컬럼을 생성 

tr_mg_2["payment_month"] = tr_mg_2["payment_date"].dt.strftime("%Y-%m")
```
