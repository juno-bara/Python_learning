```python
# 데이터 프레임의 결합 
# 단순 행, 열 결합
# 특정한 조건에 따라 열을 추가하는 경우


import pandas as pd
```

```python
# 단순한 행을 추가
df = pd.DataFrame({
    "s1" : [1,2,3,4],
    "s2" : [5,6,7,8]
})


```

```python
df_2 = pd.DataFrame([1,2])
```

```python
# concat()
# 1. 데이터프레임을 결정하는 함수
# 2. 매개변수 axis > 행을 추가할지 열을 추가할지 결정
# 3. 단순한 결합시 사용
# 4, 매개변수 ignore_index > 기본값이 false, 인덱스의 값을 그대로 유지할지 결정
```

```python
# 행을 추가 

# pd.concat([df, df_2],axis = 0,ignore_index = True)
pd.concat([df,df_2],axis = 0).reset_index(drop=True)
```

```python
df_3 = pd.DataFrame({
    "s1" : [1,2]
})
```

```python
pd.concat([df,df_3],axis = 0)
```

```python
pd.concat([df,df_3],axis = 1)
```

```python
# merge
# 1. 데이터프레임을 결합하는 함수
# 2. 특정 조건에 맞춰 열을 추가하는 함수
# 3. 매개변수 on >> 조건 
# 4. 매개변수 how >> (left, right, inner)방법
```

```python
df_4 = pd.DataFrame({
    "id" : [1,2,3,4,5],
    "item" : ["a", "b", "a", "a", "b"]
})


df_5 = pd.DataFrame({
    "id" : [1,4,5,2,3],
    "name" : ["kim", "lee", "park", "choi", "ha"]

})
```

```python
# df_4, df_5 결합
# 1. 단순히 열을 추가하는 concat

pd.concat([df_4,df_5],axis=1)



```

```python
pd.concat([df_4,df_5.sort_values("id")],axis=1)
```

```python
# 단순한 열 결합이 아닌 특정한 조건에 맞는 열 결합
# merge([데이터프레임 들],on=조건,how=방법)

pd.merge(df_4,df_5, on = "id", how = "left")
```

```python
pd.merge(df_4,df_5, on = "id", how = "right")
```

```python
pd.merge(df_4,df_5, on = "id", how = "inner")



# left right inner 순서만 다르고 같은 값이 출력됨
# df_4 df_5 모두 id 값을 같이 가지고 있기 때문
```

```python
df_6 = pd.DataFrame({
    "item" : ["a", "b", "c"],
    "item_price" : [100, 200, 300]
})
```

```python
df_6
df_4
```

```python
# df_4, df_6를 merge를 이용해서 결합
# on이라는 조건에 들어가야될 값 >> item

pd.merge(df_4,df_6,on="item", how= "left" )
```

```python

pd.merge(df_4,df_6,on="item", how= "right")
```

```python
df_7 = pd.DataFrame({
    "id" : [1,2,3,4,5,6],
    "item" : ["a", "a", "b", "a", "b", "d"]
})
```

```python
pd.merge(df_7,df_6, on="item", how="left")
```

```python
pd.merge(df_7,df_6, on="item", how="right")
```

```python
pd.merge(df_7,df_6, on="item", how="inner")
```

```python
pd.merge(df_7,df_6, on="item", how="outer")
```
