```python
kospi = fdr.DataReader("kS11", "2019")
kospi.head(3)
```

```python
# 그래프의 사이즈 조절 
plt.figure(figsize=(20,20))
# grid(눈금)추가
plt.grid(True,axis="y")
plt.grid(True,axis="x")
plt.plot(kospi["Close"], marker = "o")
plt.xlabel("날짜")
plt.ylabel("종가")
plt.show()
```

```python
# 바형 그래프 
    # bar()함수를 이용하여 그래프를 출력
    # bar(x값, y값)
    # x값이 존재하지 않으면 에러
```

```python
plt.bar([1,2,3,4],[4,3,2,1])
plt.show()
```

```python
mean_kospi_2022 = kospi.loc["2022"].resample("MS").mean()
mean_kospi_2022
```

```python
# x축은 년-월로 표시
# y축은 평균 거래량 
# 바그래프  
```

```python
x = []
for i in mean_kospi_2022.index:
    x.append(i.strftime(format="%Y년%m월"))

# mean_kospi_2022.index.strftime(format= "%Y년 %m월")

```

```python
plt.figure(figsize=(20,20))
plt.barh(x,mean_kospi_2022["Volume"])
# 지수 표현식을 일반 표현식으로 변환
plt.ticklabel_format(axis="x", style="plain")
plt.show()
```

```python
# kospi 에서 변동률 파생변수 생성
# 변동률 = (종가 - 시가) / 시가

kospi["변동률"] = (kospi["Close"] - kospi["Open"] )/ kospi["Open"] * 100
kospi.head(3)
```

```python
plt.rc("axes", unicode_minus = False)
plt.grid(True,axis="x")
plt.grid(True,axis="y")
plt.hist(kospi["변동률"], bins=20)
plt.show()
```
